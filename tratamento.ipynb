```python
import pandas as pd
import re

df = pd.read_excel("Placas_de_Video.xlsx")

display(df)
```

```python
display(df.info())
```

```python
df["preco"] = df["preco"].str.replace(r"[A-Za-zÀ-ÖØ-öø-ÿ-$]", "", regex=True)
df["parcela"] = df["parcela"].str.replace("À vista no PIX\nou até\n","")
df["avaliacoes"] = df["avaliacoes"].str.replace(r"[(-)]", "", regex=True)
df["avaliacoes"].fillna(0, inplace=True)

display(df.info())
```

```python
df['memoria'] = (
    df['marca']
      .str.extract(r'(\d+\s*[Gg][Bb])', flags=re.IGNORECASE, expand=False)
      .str.replace(r'\s+', '', regex=True)
      .str.upper()
)

df['tipo_memoria'] = (
    df['marca']
      .str.extract(r'((?:GDDR|DDR)\d|SSD)', flags=re.IGNORECASE, expand=False)
      .str.upper()
)

df['bits'] = (
    df['marca']
      .str.extract(r'(\d+\s*bits)', flags=re.IGNORECASE, expand=False)
      .str.replace(r'\s+', '', regex=True)
      .str.lower()
)


display(df)
```

```python
df['marca'] = df['marca'].str.lower()
marcas = ['asrock','gigabyte','xfx','palit','galax','nvidia','amd','msi', 'afox', 'xfx']

pattern = r'(' + '|'.join(marcas) + r')'

df['marca'] = df['marca'].str.extract(pattern, expand=False)

df['desconto'] = df['desconto'].str.replace(r'%', "")
df['desconto'] = df['desconto'].str.replace(r'-', "")
df['desconto'] = df['desconto'].astype(int)

display(df)
```

```python
df["marca"].fillna("Sem marca", inplace=True)
df["preco"] = df["preco"].str.strip()
df['preco'] = df['preco'].str.replace(r'.', "")
df['preco'] = df['preco'].str.replace(r',', ".")
df['preco'] = df['preco'].astype("float")
df[['parcelas', 'valor_un_parcela']] = df['parcela'].str.extract(r'(\d+)x de R\$ ([\d\.,]+)')

df = df.drop(columns="parcela")

display(df)
display(df.info())
```

```python
df['memoria'] = df['memoria'].str.replace(r'GB', "")
df["memoria"] = df["memoria"].fillna(0)
df["memoria"] = df['memoria'].astype("int")
display(df)
```

```python
df['valor_un_parcela'] = df['valor_un_parcela'].str.replace(r',', ".")
df["parcelas"] = df["parcelas"].astype("int")
df["valor_un_parcela"] = df["valor_un_parcela"].astype("float")
display(df.info())
```

```python
df['bits'] = df['bits'].str.replace(r'bits', "")
df['bits'] = df['bits'].fillna(0)
df['bits'] = df['bits'].astype("int")

df['marca'] = df['marca'].str.upper()

df['tipo_memoria'] = df['tipo_memoria'].fillna("Sem Tipo")

df = df.rename(columns={'memoria': 'memoria_em_GB'})
df = df.rename(columns={'desconto': 'desconto_em_%'})

display(df)
```

```python
df.to_excel("Placa_de_video_tratada.xlsx", index=False)
```
